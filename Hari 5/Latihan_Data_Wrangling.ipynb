#1. Impor library dan dataset

```python
#import pandas library
import pandas as pd

#read data
df = pd.read_csv('https://raw.githubusercontent.com/audit-ti/pjj-pengolahan-data-python/main/04.%20Data%20Preparation/dataset/sales_data_latihan.csv')
df
```

#2. Perbaikan format data

##2.1 Kolom sales seharusnya numerik (hapus Rp)

```python
df['Sales'] = df.Sales.str.replace('Rp','')
df['Sales'] = df.Sales.str.replace(',','')
df['Sales'] = df.Sales.astype('float')
df
```

##2.2 Kolom "Percent Growth" seharusnya numerik (tanpa tanda %)

```python
df['Percent Growth'] = df['Percent Growth'].str.replace('%','')
df['Percent Growth'] = df['Percent Growth'].astype('float')
df
```

##2.3 Hilangkan spasi berlebih pada kolom Type

```python
df['Type'] = df['Type'].str.strip()
df
```

#3. Buat Laporan

##3.1 Jumlah Sales Berdasarkan Tipe

```python
df[['Type','Sales']].groupby('Type').sum('Sales')
```

##3.2 Ambil 3 customer dengan kenaikan tertinggi

```python
df.sort_values('Percent Growth', ascending=False).head(3)
```
